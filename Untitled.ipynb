```python
# Import necessary libraries
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk, RegexpParser
from nltk.chunk import conlltags2tree, tree2conlltags
import string
import re
```

```python
# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('wordnet')
```

```python
# Read the text file
with open('text.txt', 'r') as file:
    text = file.read()

print("Original Text:")
print(text)
```

```python
# Remove punctuation
text_no_punctuation = text.translate(str.maketrans('', '', string.punctuation))

print("\nText After Removing Punctuation:")
print(text_no_punctuation)
```

```python
# Tokenize the text
tokens = word_tokenize(text_no_punctuation)

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\nText After Removing Stop Words:")
print(' '.join(filtered_tokens))
```

```python
# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print("\nText After Stemming:")
print(' '.join(stemmed_tokens))
```

```python
# POS Tagging
pos_tags = pos_tag(filtered_tokens)

print("\nText After POS Tagging:")
print(pos_tags)
```

```python
# Lemmatizing
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nText After Lemmatizing:")
print(' '.join(lemmatized_tokens))
```

```python
# Chunking
chunked = ne_chunk(pos_tags)

print("\nText After Chunking:")
print(chunked)

# Converting the chunked tree to IOB format for chinking demonstration
iob_tagged = tree2conlltags(chunked)

# Define a chinking pattern to remove some parts of the chunk
grammar = r"""
    NP: {<DT>?<JJ>*<NN>}  # Chunk sequences of DT, JJ, NN
    }<VB.?|IN>+{          # Chink sequences of VB* or IN
    """
chunk_parser = RegexpParser(grammar)
chinked = chunk_parser.parse(pos_tags)

print("\nText After Chinking:")
print(chinked)
```

```python
# Named-Entity Recognition
ner_tree = ne_chunk(pos_tags)

print("\nText After Named-Entity Recognition:")
print(ner_tree)
```

```python
# Rule-Based Matching using regular expressions
pattern = re.compile(r'\b\w{6}\b')  # Example: Find all 6-letter words
matches = pattern.findall(text)

print("\nText After Rule-Based Matching (6-letter words):")
print(matches)
```

```python
# Bag of Words
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform([text])
print("\nBag of Words:")
print(bag_of_words.toarray())
```

```python
import spacy
from spacy.matcher import Matcher, PhraseMatcher
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# Load SpaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# Read the text file
with open('text.txt', 'r') as file:
    text = file.read()
```

```python
# Process the text
doc = nlp(text)
```

```python
# Remove punctuation
text_no_punctuation = " ".join([token.text for token in doc if not token.is_punct])

print("Text After Removing Punctuation:")
print(text_no_punctuation)
```

```python
# Remove stop words
filtered_tokens = [token.text for token in doc if not token.is_stop]

print("\nText After Removing Stop Words:")
print(" ".join(filtered_tokens))
```

```python
# Lemmatization
lemmatized_tokens = [token.lemma_ for token in doc]

print("\nText After Lemmatizing:")
print(" ".join(lemmatized_tokens))
```

```python
# Part of Speech Tagging
pos_tags = [(token.text, token.pos_) for token in doc]

print("\nText After POS Tagging:")
print(pos_tags)
```

```python
# Named-Entity Recognition
entities = [(ent.text, ent.label_) for ent in doc.ents]

print("\nNamed-Entity Recognition:")
print(entities)
```

```python
# Dependency Parsing
dependencies = [(token.text, token.dep_, token.head.text) for token in doc]

print("\nDependency Parsing:")
print(dependencies)
```

```python
# Rule-Based Matching
matcher = Matcher(nlp.vocab)
# Define patterns
pattern_digit = [{"IS_DIGIT": True}]  # Match digits
pattern_spacy = [{"LOWER": "muslims"}]  # Match the word 'muslims' in lowercase
pattern_phrase = [{"LOWER": "holy"}, {"LOWER": "qur'an"}]  # Match the phrase 'holy qur'an'

# Add patterns to the matcher
matcher.add("DigitPattern", [pattern_digit])
matcher.add("MuslimsPattern", [pattern_spacy])
matcher.add("HolyQuranPattern", [pattern_phrase])

# Find matches
matches = matcher(doc)

print("\nRule-Based Matching:")
for match_id, start, end in matches:
    span = doc[start:end]
    match_label = nlp.vocab.strings[match_id]  # Get the string representation of the match_id
    print(f"{match_label}: {span.text}")

# Check tokens and their attributes (for debugging purposes)
print("\nTokens and their attributes:")
for token in doc:
    print(f"Token: {token.text}, IS_DIGIT: {token.is_digit}, LOWER: {token.lower_}, LEMMA: {token.lemma_}, POS: {token.pos_}")
```

```python
# Word2Vec (pre-trained vectors in SpaCy)
print("\nWord2Vec (pre-trained vectors):")
for token in doc:
    print(token.text, token.vector[:5])  # print first 5 elements of the vector
```
